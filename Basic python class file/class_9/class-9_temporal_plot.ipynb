# Multiple bar charts
 
based on thickness and the positions of the bars

thickness = 0.25 units

Each bar chart will be shifted 0.25 units from the previous one.

```python
import numpy as np

English    = [60, 75, 80, 30]
Mathmetics = [17, 20, 50, 45]
Physics    = [70, 50, 75, 30]

x = np.arange(4)  # create positions

plt.bar(x + 0.00, English, color = 'b', width = 0.25, label = 'English')
plt.bar(x + 0.25, Mathmetics, color = 'r', width = 0.25, label = 'Mathmetics')
plt.bar(x + 0.50, Physics, color = 'g', width = 0.25, label = 'Physics')
plt.legend()
plt.xticks([0 + 0.25, 1 + 0.25, 2 + 0.25, 3 + 0.25],[2017, 2018, 2019, 2020])  # ticks, labels
```

# Stacked bar chart 

Instead of running from zero to a value, it will go from the bottom to the value.

```python
plt.bar(x, English, color = 'b', width = 0.25, label = 'English')
plt.bar(x, Mathmetics, color = 'r', width = 0.25, bottom = English, label = 'Mathmetics')
plt.legend()
plt.xticks([0, 1, 2, 3],[2017, 2018, 2019, 2020])  # ticks, labels
```

# Subplot

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot.html
    
subplot(nrows, ncols, index, **kwargs)

default: (1, 1, 1)
    
index starts at 1 in the upper left corner and increases to the right.

```python
plt.subplot(211)    
plt.plot(data['year'],data['pet'], color='green', marker='o', linestyle='dashed', linewidth=2, markersize=10)

plt.subplot(212)
plt.plot(data1['year'],data1['di'], color='red', marker='o', linestyle='dashed', linewidth=2, markersize=10)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1)

# make a little extra space between the subplots
fig.subplots_adjust(hspace=0.5)

ax1.plot(data['year'],data['pet'], color='green', marker='o', linestyle='dashed', linewidth=2, markersize=10)
ax1.set_xlabel('Year')
ax1.set_ylabel('PET (mm)')

ax2.plot(data1['year'],data1['di'], color='red', marker='o', linestyle='dashed', linewidth=2, markersize=10)
ax2.set_xlabel('Year')
ax2.set_ylabel('DI (mm)')

plt.savefig("PET_DI.png",dpi=300)
plt.show()
```

```python
fig, ax1 = plt.subplots(1, 1, figsize=(10,4), dpi= 80)

ax1.plot(data['year'],data['pet'], color='green', marker='o', linestyle='dashed', linewidth=2, markersize=10, label = 'PET')
ax1.set_xlabel('Year')
ax1.set_ylabel('PET (mm)')
ax1.legend(loc='upper center')


# Plot Line2 (Right Y Axis)
ax2 = ax1.twinx()  # Make and return a second axes that shares the x-axis.

ax2.plot(data1['year'],data1['di'], color='red', marker='o', linestyle='dashed', linewidth=2, markersize=10, label = 'DI')
ax2.set_xlabel('Year')
ax2.set_ylabel('DI (mm)')
ax2.legend(loc='upper right')
fig.tight_layout()

plt.savefig("secondary_Y_axis.png",dpi=300)
plt.show()
```

# Contour plots

 Z = f(X,Y)   
    
how alue Z changes as a function of two inputs X and Y

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

msw = pd.read_excel("MSW.xlsx")
msw.head()
```

```python
msw.shape
```

```python
latitude  = msw['Latitude (N)']
longitude = msw['Longitude (E)']
MSW       = msw['MSW (ms-1)']

              
lat, lon      = np.meshgrid(latitude, longitude)
MSW_ms, lat_  = np.meshgrid(MSW, latitude)

plt.contourf(lon, lat, MSW_ms)    # contourf (X, Y, Z)
plt.colorbar()     # Add a colorbar to a plot
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.xlim(140,152)
plt.ylim(8,36)
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax  = plt.axes(projection='3d')
ax.contour3D(lon, lat, MSW_ms, 50, cmap='viridis')
ax.set_xlabel('longitude')
ax.set_ylabel('latitude')
ax.set_zlabel('Maximum sustained wind speed (m/s)')
plt.show()
```
