# Create a series in Pandas

```python
x = np.arange(0,6,1)
x
```

```python
y = list(x)
y
```

```python
ser = pd.Series(y)
print(ser)
```

# Use of .loc and .iloc

.loc locates by the index name (select by label)

.iloc locates by numerical index (select by position)


```python
data1.loc[0:5,['year','pet']]  # Here, we are selecting only 6 years "pet" data.
```

```python
data1.iloc[1]
```

```python
data1.iloc[0,1]
```

```python
data1.iloc[0:6,0:6]
```
